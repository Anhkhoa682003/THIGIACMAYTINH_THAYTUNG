```python
# Transpose ảnh
from PIL import Image
from imgtools import *
from IPython.display import display # to display images
```

```python
# Đường dẫn ảnh
dir = 'C:/Users/Acer/Pictures'
```

```python
img_path = dir + '/211.PNG'
```

```python
#In đường dẫn ảnh
print(img_path)
```

```python
#Đọc ảnh
img = Image.open(img_path)
```

```python
display(img)
```

```python
# Chuyển từ ảnh RGB sang ảnh xám
gray_img = img.convert("L")
```

```python
display(gray_img)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band, _ = img.split()
```

```python
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
```

```python
display(merged_image)
```
