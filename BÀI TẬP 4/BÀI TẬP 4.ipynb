```python
# Import lớp Image
from PIL import Image
from imgtools import *
from IPython.display import display
```

```python
# Đường dẫn ảnh
dir = 'C:/Users/Acer/Pictures'
```

```python
image_path = dir +'/211.PNG'
```

```python
#In đường dẫn ảnh
print(image_path)
```

```python
# Đọc ảnh 
img01 = Image.open(image_path)
```

```python
display(img01)
```

```python
# Định vị khu vực cắt
region = (100, 75, 175, 160)
# Cắt vùng ảnh
cropped_image = img01.crop(region)
display(cropped_image)
```

```python
# Xác định vị trí cần dán và dán ảnh
paste_position = (150, 150)
img01.paste(cropped_image, paste_position)
display(img01)
```
