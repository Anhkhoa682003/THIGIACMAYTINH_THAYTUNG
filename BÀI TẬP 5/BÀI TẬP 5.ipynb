```python
# Import lớp Image
from PIL import Image
from imgtools import *
from IPython.display import display # to display images
```

```python
# Đường dẫn ảnh
dir = 'C:/Users/Acer/Pictures'
```

```python
img_path = dir + '/211.PNG'
```

```python
#In đường dẫn ảnh
print(img_path)
```

```python
#Đọc ảnh
img = Image.open(img_path)
```

```python
display(img)
```

```python
#Xoay ảnh
rotated_img = img.rotate(90)
```

```python
display(rotated_img)
```

```python
# Transpose ảnh
transposed_img = img.transpose(Image.Transpose.ROTATE_90)
```

```python
display(transposed_img)
```
