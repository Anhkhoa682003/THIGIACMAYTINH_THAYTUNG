```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
dir = 'C:/Users/Acer/Pictures'
```

```python
image_path = dir +'/211.PNG'
```

```python
#In đường dẫn ảnh
print(image_path)
```

```python
img01 = Image.open(image_path)
```

```python
plt.imshow(img01)
```

```python
# Chuyển sang ảnh xám
img01_gray = img01.convert('L')
plt.imshow(img01_gray, cmap='gray')
```

```python
img01_array = np.array(img01_gray)
```

```python
print(img01_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(img01_array.flatten(), bins=128)
plt.title('BIỂU ĐỒ HISTOGRAM CỦA ẢNH')
plt.xlabel('GIÁ TRỊ CỦA PIXEL')
plt.ylabel('SỐ LƯỢNG')
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(img01_gray, origin='image')
plt.show()
```
